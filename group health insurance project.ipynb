In simple words, outliers are extremely high or low values in a dataset. Outliers influence the natural distribution of the data and can synthetically inflate or deflate the mean. They can exist due to many genuine reasons. To identify the outliers and treat them is a critical task in the process of data cleansing. Visualization techniques are the best when it comes to identifying the outliers. Boxplots help identify outliers in the data easily.

# Feature Engineering

```python
#
outliers treatment using IQR

def find_boundaries(df, variable):
    Q1 = df[variable].quantile(.25)
    Q3 = df[variable].quantile(.75)
    IQR= Q3-Q1
    lower_boundary=Q1-1.5*IQR
    upper_boundary=Q3+1.5*IQR
    #Extreme Outliers
    lower_bridge1=Q1-(IQR*3)
    upper_bridge1=Q3+(IQR*3)
    return lower_boundary,upper_boundary, lower_bridge1,upper_bridge1

```

```python
lower,upper,ext_lower,ext_upper = find_boundaries(data,'Group_Size')
print(lower,upper,ext_lower,ext_upper)

data.loc[data['Group_Size']>212,'Group_Size'] = 212
```

```python
lower,upper,ext_lower,ext_upper = find_boundaries(data,'Sum_Insured')
print(lower,upper,ext_lower,ext_upper)

data.loc[data['Sum_Insured']>800000.0,'Sum_Insured'] = 800000.0
```

```python
lower,upper,ext_lower,ext_upper = find_boundaries(data,'Premium_per_Year')
print(lower,upper,ext_lower,ext_upper)

data.loc[data['Premium_per_Year']>2473211.516918265,'Premium_per_Year'] = 2473211.0
```

```python
lower,upper,ext_lower,ext_upper = find_boundaries(data,'Ambulance_Charges')
print(lower,upper,ext_lower,ext_upper)

data.loc[data['Ambulance_Charges']>3500,'Ambulance_Charges'] = 3500
```

```python
lower,upper,ext_lower,ext_upper = find_boundaries(data,'Network_Hospitals')
print(lower,upper,ext_lower,ext_upper)

data.loc[data['Network_Hospitals']>10750.0 ,'Network_Hospitals'] = 10750.0 
```
